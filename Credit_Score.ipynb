```python
df1=df.drop(["Gender"],axis=1)
df1
```

```python
df1["Income"].fillna(df["Income"].mean(),inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1['Education']=le.fit_transform(df1['Education'])
df1['Marital Status']=le.fit_transform(df1['Marital Status'])
df1['Home Ownership']=le.fit_transform(df1['Home Ownership'])
df1['Credit Score']=le.fit_transform(df1['Credit Score'])
```

```python
x=df1.iloc[:,:-1].values
x
```

```python
y=df1.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
knn=KNeighborsClassifier(n_neighbors=7)
base=GaussianNB()
model=SVC()
lst_model=[knn,base,model]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
for i in lst_model:
  print(i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("********************************************************************************")
  print(classification_report(y_test,y_pred))
  print("********************************************************************************")
  print(confusion_matrix(y_test,y_pred))
  print("********************************************************************************")
  print(classification_report(y_test,y_pred))
```
